```python
plt.clf()

#  read x-y points to plot
x,y = [], []
f=open('ch150sa','r') 
for l in f:
    row = l.split()
    x.append(row[0])
    y.append(row[1])
x = [int(i) for i in x]
y = [float(i) for i in y]
    
plt.title('Simulated Annealing')

plt.xlabel('Iteration') # naming the x axis 
plt.ylabel('Tour Length')          # naming the y axis :
plt.style.use('default')


plt.plot(x, y, color='red', linestyle='dashed', linewidth = 1.00, 
         marker='o', markerfacecolor='green', markersize=5) 
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))

#plt.show()

#plt.show()
plt.savefig('sa_charts.png',dpi=300)

```

```python
# complete set up for line plot

def make_plot(filename, titlename) :
    #  read x-y points to plot
    x,y = [], []
    f=open(filename,'r') 
    for l in f:
        row = l.split()
        x.append(row[0])
        y.append(row[1])
    
    # naming the x axis 
    plt.xlabel('Number of Iterations') 
    # naming the y axis 
    plt.ylabel('Tour Length') 

    plt.style.use('default')

    plt.plot(x, y, color='red', linestyle='dashed', linewidth = 0.5, 
         marker='o', markerfacecolor='green', markersize=5) 

    plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))

    plt.show()
```

```python
make_plot('ch150sa','Simulated Annealing')
```
