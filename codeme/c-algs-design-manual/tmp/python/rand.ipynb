```python
import random
import matplotlib.pyplot as plt
```

```python
def sim():
    arr = []
    for _ in range(16):
        arr.append(random.random() + 0.4)
    total = sum(arr)
    for i in range(16):
        arr[i] /= total
    b = arr[4:8]
    c = arr[12:16]
    b = [x*7 for x in b]
    c = [x*7 for x in c]
    d = arr[:4] + b + arr[8:12] + c
    total = sum(d)
    for i in range(16):
        d[i] /= total
    x = []
    for i in range(16):
        r = bin(i)[2:]
        r = "0"*(4-len(r)) + r
        x.append(r)
    plt.xticks(rotation=90)
    plt.ylim(0,0.2)
    plt.plot(x,arr,label="pre-jack")
    plt.plot(x,d,label="post-jack")
    plt.legend()
```

```python
sim()
```
